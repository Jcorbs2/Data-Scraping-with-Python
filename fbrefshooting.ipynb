```python
from unidecode import unidecode
import pandas as pd
import numpy as np
import requests
from io import StringIO
```

```python
url = 'https://fbref.com/en/comps/Big5/shooting/players/Big-5-European-Leagues-Stats'
html_content = requests.get(url).text.replace('<!--', ' ').replace('-->', ' ')
df=pd.read_html(StringIO(html_content))
```

```python
dfshooting = dfdata.drop(dfdata[dfdata.Age == 'Age'].index)
```

```python
dfshooting.shape
```

```python
dfshooting.columns
```

```python
dfshooting['90s'] = dfshooting['90s'].astype(float)
dfshooting['Gls'] = dfshooting['Gls'].astype(float)
dfshooting['Sh'] = dfshooting['Sh'].astype(float)
dfshooting['SoT'] = dfshooting['SoT'].astype(float)
dfshooting['SoT%'] = dfshooting['SoT%'].astype(float)
dfshooting['G/Sh'] = dfshooting['G/Sh'].astype(float)
dfshooting['G/SoT'] = dfshooting['G/SoT'].astype(float)
dfshooting['Dist'] = dfshooting['Dist'].astype(float)
dfshooting['FK'] = dfshooting['FK'].astype(float)
dfshooting['PK'] = dfshooting['PK'].astype(float)
dfshooting['PKatt'] = dfshooting['PKatt'].astype(float)
dfshooting['xG'] = dfshooting['xG'].astype(float)
dfshooting['npxG'] = dfshooting['npxG'].astype(float)
dfshooting['npxG/Sh'] = dfshooting['npxG/Sh'].astype(float)
dfshooting['G-xG'] = dfshooting['G-xG'].astype(float)
dfshooting['np:G-xG'] = dfshooting['np:G-xG'].astype(float)
```

```python
dfshooting['Gls'] = dfshooting['Gls'] / dfshooting['90s']
dfshooting['Sh'] = dfshooting['Sh'] / dfshooting['90s']
dfshooting['SoT'] = dfshooting['SoT'] / dfshooting['90s']
dfshooting['G/Sh'] = dfshooting['G/Sh'] / dfshooting['90s']
dfshooting['G/SoT'] = dfshooting['G/SoT'] / dfshooting['90s']
dfshooting['Dist'] = dfshooting['Dist'] / dfshooting['90s']
dfshooting['FK'] = dfshooting['FK'] / dfshooting['90s']
dfshooting['PK'] = dfshooting['PK'] / dfshooting['90s']
dfshooting['PKatt'] = dfshooting['PKatt'] / dfshooting['90s']
dfshooting['xG'] = dfshooting['xG'] / dfshooting['90s']
dfshooting['npxG'] = dfshooting['npxG'] / dfshooting['90s']
dfshooting['npxG/Sh'] = dfshooting['npxG/Sh'] / dfshooting['90s']
dfshooting['G-xG'] = dfshooting['G-xG'] / dfshooting['90s']
dfshooting['np:G-xG'] = dfshooting['np:G-xG'] / dfshooting['90s']
```

```python
dfshooting['PlSqu'] = dfshooting['Player'] + dfshooting['Squad']
```

```python
dfshooting['Player'] = dfshooting['Player'].apply(unidecode)
```

```python
dfshooting['Squad'] = dfshooting['Squad'].apply(unidecode)
```
